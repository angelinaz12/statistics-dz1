```python
#считаем выбросы для каждой колокни
def vibros(column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df8[df8[column].between(lower_bound, upper_bound, inclusive=True)].sort_values(column)
    vibrosy = pd.concat([df8, remove_outliers]).drop_duplicates(keep=False)
    return vibrosy
```

```python
for column in df8:
    display(vibros(column))
```
